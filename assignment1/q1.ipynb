# Importing libraries

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

# KNNClassifer

```python
class KNNClassifier():
    train_data = np.empty(2)
    validation_data = np.empty(2)
    test_data = np.empty(2)
    train_data1 = np.empty(2)
    k = 3
    def _init__():
        print("welcome")
        
    def euclidean(self,a, b):
        dist = np.linalg.norm(a-b)
        return dist 

    def manhattan(self,ippt,testpt):
        sum_ = 0
        for i in range(len(testpt)):
            sum_ += abs(ippt[i] - testpt[i])

        return sum_
    
    def mode(self,labels):
        return Counter(labels).most_common(1)[0][0]
    
    
    def predict(self,test_data_path):
        self.test_data = pd.read_csv(test_data_path, header=None)
        print("shape = ",self.train_data1.shape)
        print("shape = ",self.test_data.shape)
        self.test_data = self.test_data.to_numpy()
        self.train_data1 = self.train_data1.to_numpy()
        labels = self.predict_values_for_validation(self.train_data1, self.test_data, self.k, distance_function='manhattan')
        return labels
        
    def predict_values_for_validation(self, train_data, test_data, k, distance_function):
        final_labels = []
        for j in range(len(test_data)):
            distance_labels_and_indexes = []
            labels = []
            dist = 0
            for i in range(len(train_data)):
                dist = self.euclidean(train_data[i,1:], test_data[j,:])
                distance_labels_and_indexes.append((dist, i))
            sorted_dist = sorted(distance_labels_and_indexes)
            first_k_values = sorted_dist[:k]
            for dist,i in first_k_values:
                index = i
                labels.append(train_data[i][0])
            label =  self.mode(labels)
            final_labels.append(label)
        return final_labels
    
    def predict_values_for_validation_for_manhattan(self, train_data, test_data, k, distance_function):
        final_labels = []
        for j in range(len(test_data)):
            distance_labels_and_indexes = []
            labels = []
            dist = 0
            for i in range(len(train_data)):
                dist = self.manhattan(train_data[i,:], test_data[j,:])
                distance_labels_and_indexes.append((dist, i))
            sorted_dist = sorted(distance_labels_and_indexes)
            first_k_values = sorted_dist[:k]
            for dist,i in first_k_values:
                index = i
                labels.append(train_data[i][0])
            label =  self.mode(labels)
            final_labels.append(label)
        return final_labels
        
        
        
    def train(self,train_path):
        self.train_data1 = pd.read_csv(train_path)
        print(self.train_data1.shape)
        '''only for validation purpose'''
        self.train_data = (self.train_data1.iloc[:19900,:]).to_numpy()
        train_data_labels = self.train_data1.iloc[:19900,0].to_numpy()
        print(self.train_data.shape)
        self.validation_data = (self.train_data1.iloc[19900:,1:]).to_numpy()
        labels_valid = self.train_data1.iloc[19900:,0].to_numpy()
        print(self.validation_data.shape)
        knn_train_data = self.train_data1.iloc[:19900, 1:].to_numpy()
        accuracies = []
        for i in range(1,6,2):
            
            self.k = i
            neigh = KNeighborsClassifier(n_neighbors=i)
            print("self.k for euclidean  = ",self.k)
            accuracy_euclidean = self.predict_values_for_validation(self.train_data, self.validation_data, self.k  , distance_function = 'euclidean') 
            print("accuracy: ",accuracy_euclidean)
            print(classification_report(labels_valid, accuracy_euclidean))
            accuracies.append(accuracy_score(labels_valid, accuracy_euclidean))
            print(confusion_matrix(labels_valid, accuracy_euclidean))
            
            print("self.k for manhattan  = ", self.k)
            accuracy_manhattan = self.predict_values_for_validation_for_manhattan(self.train_data, self.validation_data, self.k  , distance_function = 'manhattan') 
            print("accuracy: ",accuracy_manhattan)
            print(classification_report(labels_valid, accuracy_manhattan))
            accuracies.append(accuracy_score(labels_valid, accuracy_manhattan))
            print(confusion_matrix(labels_valid, accuracy_manhattan))
            print("Sklearn KNN Classifier")
            neigh.fit(knn_train_data, train_data_labels)
            print("train", self.train_data.shape)
            print(train_data_labels.shape)
            print(self.validation_data.shape)
            values = neigh.predict(self.validation_data)
            print("accuracy = ",accuracy_score(labels_valid, values))
        return accuracies
```
